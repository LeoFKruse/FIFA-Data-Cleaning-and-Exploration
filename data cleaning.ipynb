```python
# Import der relevanten Bibliotheken
import pandas as pd
import numpy as np
```

```python
# Import des Datensatzes
fifa_data = pd.read_csv(r'C:\Users\LK\OneDrive\Documents\PY Projects\Fifa Projekt\fifa21 raw data v2.csv')
fifa_data
```

```python
# 2. Schritt ~ Data Cleaning
# Unnötige Spalten löschen
fifa_data.drop(columns=['photoUrl','Name','playerUrl','Loan Date End'], inplace=True)
fifa_data.head()
```

```python
# Umbennenung der Spalten
fifa_data.rename(columns= {'LongName' : 'Name',
                           '↓OVA' : 'Overall',
                           'POT' : 'Potential',
                           'BOV' : 'Best Overall',
                           }, inplace=True)
fifa_data.head()
```

```python
# Zeilenumbruch aus der Spalte "Club" entfernen
fifa_data['Club'] = fifa_data['Club'].str.replace('\n\n\n\n','')
fifa_data.head()
```

```python
# Aufteilung der Spalte "Contract" in "Start of Contract" und "End of Contract"
fifa_data[['Start of Contract', 'End of Contract']] = fifa_data['Contract'].str.split('~', expand=True)
fifa_data['Start of Contract'] = fifa_data['Start of Contract'].str.strip() 
fifa_data['End of Contract'] = fifa_data['End of Contract'].str.strip()

# Entfernen der Spalte "Contract"
fifa_data.drop('Contract', axis=1, inplace=True)
fifa_data.head()
```

```python
# Umwandlung der Spalten "Weigth" und "Heigth" in Integers
fifa_data['Height'].unique()
```

```python
# Entfernen der Größeneinheit "cm"
fifa_data['Height'] = fifa_data['Height'].str.removesuffix('cm')
fifa_data.head()
```

```python
# Funktion die Fuß/Zoll in Zentimeter umrechnet
def feet_to_cm(height):
    if "'" in height:
        feet, inches = height.split("'")
        inches = inches.replace('"','')
        cm = float(feet) * 30.48 + float(inches) * 2.54
        return int(round(cm, 0))
    else:
        return int(height)

# Anwendung der Funktion auf die Spalte
fifa_data['Height'] = fifa_data['Height'].apply(feet_to_cm)
```

```python
# Überprüfung ob die Funktion die gewünschte Veränderung gebracht hat
fifa_data['Height'].unique()
```

```python
# Überprüfung der Spalte auf die individuellen Werte
fifa_data['Weight'].unique()
```

```python
# Entfernen der Beschriftungen und Umwandlung von Pfund in Kilogramm
fifa_data['Weight'] = fifa_data['Weight'].str.removesuffix('kg')
def lbs_to_kg(weight):
    if "lbs" in weight:
        weight = weight.replace('lbs','')
        kilo = float(weight) * 0.45
        return int(round(kilo,0))
    else:
        return int(weight)
    
# Anwendung der Funktion auf die Spalte
fifa_data['Weight'] = fifa_data['Weight'].apply(lbs_to_kg) 
```

```python
# Überprüfung ob die Funktion die gewünschte Veränderung gebracht hat
fifa_data['Weight'].unique()
```

```python
# Entfernen der Sterne aus den Spalten "W/F", "SM" und "IR"
spalten = ['W/F','SM','IR']
for spalte in spalten:
    fifa_data[spalte] = fifa_data[spalte].astype(str)
    if " " in fifa_data[spalte]:
        fifa_data[spalte] = fifa_data[spalte].str.removesuffix(' ★')
    else:
        fifa_data[spalte] = fifa_data[spalte].str.removesuffix('★')
    fifa_data[spalte] = fifa_data[spalte].astype(int)
fifa_data.head()
```

```python
# Anpassen der Spalten 'Value','Wage' und 'Release Clause' zu numerischen Werten

# Funktion die die Werte in den Spalten in numerische Werte umwandelt
def numbers(amount):
    if '€' in amount:
        amount = amount.replace('€','')
    if "M" in amount:
        return int(float(amount.replace('M',''))*1000000)
    elif "K" in amount:
        return int(float(amount.replace('K',''))*1000)

fifa_data['Value'] = fifa_data['Value'].astype(str)
fifa_data['Wage'] = fifa_data['Wage'].astype(str)
fifa_data['Release Clause'] = fifa_data['Release Clause'].astype(str)

fifa_data['Value'] = fifa_data['Value'].apply(numbers)
fifa_data['Wage'] = fifa_data['Wage'].apply(numbers)
fifa_data['Release Clause'] = fifa_data['Release Clause'].apply(numbers)
fifa_data.head()
```

```python
# Änderung des Datums, sodass es für die Analyse der Daten genutzt werden kann

# Funktion zur Umwandelung des Monats in die entsprechend Zahl
def month_to_number(x):
    if x== 'Jan':
         return 1
    elif x == 'Feb':
        return 2
    elif x == 'Mar': 
        return 3
    elif x == 'Apr':
        return 4
    elif x == 'May':
        return 5
    elif x == 'Jun': 
        return 6
    elif x == 'Jul': 
        return 7
    elif x == 'Aug': 
        return 8
    elif x == 'Sep': 
        return 9
    elif x == 'Oct': 
        return 10
    elif x == 'Nov':  
        return 11
    elif x == 'Dec':  
        return 12
    
# Funktion zur Zusammenstellung des Datums
def date_to_number(date):
    month = month_to_number(date[:3])

    if len(date) == 11:
            day = date[4]
    elif len(date) == 12:
            day = date[4:6]
    else:
         return "Ungültiges Datum"

    year = date[-4:]

    day = day.zfill(2)

    return f"{day}/{month:02}/{year}"
    
fifa_data['Joined'] = fifa_data['Joined'].apply(date_to_number)
fifa_data.head()
```

```python
# Erstellen einer neuen Spalte die Rolle der Position beschreibt

# Überprüfung der Spalte auf die individuellen Werte
fifa_data['Best Position'].unique()
```

```python
# Erstellen einer Bibliothek um diese als Werte für die neue Spalte zu verwenden
group = {'RW': 'Forward',
          'ST': 'Forward',
            'GK': 'Goalkeeper',
         'CAM': 'Midfielder',
            'LW': 'Forward',
         'CB': 'Defender',
           'CDM': 'Midfielder',
         'CF': 'Forward',
           'CM': 'Midfielder',
         'RB': 'Defender',
         'LB': 'Defender',
         'LM': 'Midfielder',
           'RM': 'Midfielder',
         'LWB': 'Defender',
           'RWB': 'Defender'
}

# Anwenden der Funktion
fifa_data['Role'] = fifa_data['Best Position'].map(group)
fifa_data.head()
```

```python
# Hinzufügen von eines weiteren Datensatzes. Ziel dabei ist es eine Spalte mit den Kontinenten zu haben, aus denen die Spieler kommen.
continents = pd.read_csv(r'C:\Users\LK\OneDrive\Documents\PY Projects\Wetter Projekt\countries.csv')
continents.head()
```

```python
# Überprüfung der Spalte auf die individuellen Werte
continents['Region'].unique()
```

```python
# Umbennung der Kontinente, um die Lesbarkeit zu verbessern
rename = {
    'AFRICA' : 'Africa',
    'EUROPE' : 'Europe',
    'LATIN' : 'South America',
    'ASIA' : 'Asia',
    'OCEANIA' : 'Oceania',
    'NEAR_EAST' : 'Asia',
    'NORTHERN_AMERICA' : 'North America'
}
continents["Region"] = continents["Region"].map(rename)
continents.head()
```

```python
# Zusammenlegen der beiden Datensätze
data = pd.merge(fifa_data, continents, left_on='Nationality', right_on='Country', how='left')
data = data.drop(columns=['Country','Population'])
data.head()
```

```python
# 3. ~ Analyse der Daten
```

```python
# Import der relevanten Bibliotheken
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 3.1 Grundlegende Untersuchung der Daten

# Übersicht über die Herkunftskontinente der Spieler 
region_count = data['Region'].value_counts()
sns.barplot(x=region_count.index, y=region_count.values)
plt.show()
print(region_count)
```

```python
# Durschnittliches Alter der Spieler
sns.histplot(data['Age'], kde=True)
plt.show()
print('Average Age:',data['Age'].mean())
```

```python
# Durschnittliches Gehalt der Spieler
plt.figure(figsize=(14,5))
sns.stripplot(x=data['Wage'])
plt.show()
print('Average Wage:',data['Wage'].mean())
```

```python
#3.2 Tiefergehende Analyse der Daten

# In welchem Alter, ist der Marktwert durschnittlich am höchsten
plt.figure(figsize=(10,5))
sns.lineplot(x='Age', y='Value', data=data, marker='o' )
plt.xlabel('Age')
plt.ylabel('Average market value (mil.)')
plt.title('Average market value by age')
plt.show()

age_market_value_mean = data.groupby('Age')['Value'].mean().reset_index()
age_market_value_mean['Value'] = age_market_value_mean['Value']/1e6
print(age_market_value_mean)
```

```python
# Zusammenhang zwischen Alter und Gehalt
sns.scatterplot(data=data, x='Age', y='Wage')
plt.xlabel('Age')
plt.ylabel('Wage')
plt.title('Wage by Age')
plt.show()


print(data.sort_values(by='Wage', ascending=False)[['Name','Wage','Age']])
```

```python
# Gibt es bestimmte Größenunterschiede für die unterschiedlichen Positionen
plt.figure(figsize=(10, 6))
sns.boxplot(x='Best Position', y='Height', data=data, )
plt.title('Height by Role')
plt.show()

position_by_height = data.groupby('Best Position')['Height'].mean().reset_index()
print(position_by_height.sort_values(by='Height',ascending=False))
```
